# Merge engineered features into application train

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# import main application dataset

application_train = pd.read_csv('../data/application_train.csv')
```

# 1. Merge features from BUREAU dataset

```python
# import BUREAU engineered features

engineered_bureau = pd.read_csv('../data/engineered_bureau.csv')

engineered_bureau
```

```python
# left merge engineered_bureau into the main application train on SK_ID_CURR

merged_df = pd.merge(application_train, engineered_bureau, on='SK_ID_CURR', how='left')
```

```python
# check the merged dataframe

merged_df.head()
```

```python
# check for nulls and count

merged_df.info(verbose=True, show_counts=True)
```

# 2. Merge Features from BUREAU_BALANCE dataset

```python
# import engineered features from bureau_balance

engineered_bureau_balance = pd.read_csv('../data/engineered_bureau_balance.csv')

# check the imported data

engineered_bureau_balance
```

```python
# merge engineered_bureau_balance into main application data on SK_ID_CURR

merged_df = pd.merge(merged_df, engineered_bureau_balance, on='SK_ID_CURR', how='left').fillna(0)
merged_df
```

# 3. Merge Features from CREDIT_CARD_BALANCE dataset

```python
# import dataset

engineered_cc = pd.read_csv('../data/engineered_cc.csv')
engineered_cc.head()
```

```python
# merge into the engineered features dataframe

merged_df = pd.merge(merged_df, engineered_cc, on='SK_ID_CURR', how='left')

# check the merged df

merged_df
```
